```python
#4
letras = []
vogais = ['A','E','I','O','U']

for i in range(10):
    letras.append(raw_input('Digite o caracter %d: '%(i+1)).upper())

totalConsoantes = 0
consoantes = []

for letra in letras:
    if letra not in vogais:
        consoantes.append(letra)
        totalConsoantes += 1

print 'Numero de consoantes = ',totalConsoantes
print 'Consoantes = ', consoantes
```

```python
#6
notas = []
for i in range(0, 10):
    notasAluno = []
    for j in range(0, 4):
        notasAluno.append(
            float(raw_input('Informe a nota do aluno %.d: ' % (i + 1))))
    notas.append(notasAluno)

alunosMedia = 0
for notasAluno in notas:
    somaNotas = 0
    for nota in notasAluno:
        somaNotas += nota
    if ((somaNotas / 4.0) >= 7.0):
        alunosMedia += 1

print '%d alunos ficaram acima da media' % alunosMedia
```

```python
#7
numeros = []
for i in range(0, 4):
    numeros.append(int(raw_input('Informe um numero inteiro: ')))

soma = 0
multiplicacao = 1
for i in numeros:
    soma += i
    multiplicacao *= i

print numeros
print 'Soma dos numeros: %d' % soma
print 'Multiplicacao dos numeros: %d' % multiplicacao
```

```python
#8
pessoas = []
for i in range(0, 5):
    pessoa = []
    pessoa.append(raw_input('Informe a idade da pessoa: '))
    pessoa.append(raw_input('Informe a altura da pessoa: '))
    pessoas.append(pessoa)

pessoas.reverse()
for pessoa in pessoas:
    print 'Idade: %s - Altura: %s' % (pessoa[0], pessoa[1])
```

```python
#Mega sena

import random

lancamento = []
sorteados = [0]*61
recomendados = [0]*6

for i in range(10000):
    numero = random.randint(1, 60)
    lancamento.append(numero)
   
for i in range(1,61):
    sorteados[i]=(lancamento.count(i),i)
    #print '(Numero, sorteados) = ', sorteados[i]

for i in range (6):
    maximo = max(sorteados)
    indice = sorteados.index(maximo)
    recomendados[i]=sorteados[indice][1]
    sorteados.pop(indice)
    
recomendados.sort()
print 'Numeros recomendados = ', recomendados
```
