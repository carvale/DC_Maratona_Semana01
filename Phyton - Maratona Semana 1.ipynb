```python
#Zerinho ou Um
A, B, C = input().split()
ganhador = None

if A != B and A != C:
    ganhador = 'A'
elif B != A and B != C:
    ganhador = 'B'
elif C != A and C != A:
    ganhador = 'C'
elif A == B and B == C:
    ganhador = '*'

print (ganhador)
```

```python
#Premio do MilhÃ£o
N = int (input())

visualizacoes = []
for i in range (N):
    A = int(input())
    visualizacoes.append(A)

total = 0
resposta = -1

for i,v in enumerate(visualizacoes):
    dia = i + 1
    total = total + v
    if total >= 1000000 and resposta == -1:
        resposta = dia

print(resposta)

```

```python
import requests
import json

sair = False

def requisicao(titulo):
    try:
        req = requests.get("http://www.omdbapi.com/?t="+ titulo + '&type=movie'+ '&apikey=b07c19cd')
        dicionario = json.loads(req.text)
        return dicionario
    except:
        print('Erro na conexÃ£o')
        return None

def printar_detalhes(filme):
    print('TÃ­tulo', filme['Title'])
    print('Ano:', filme['Year'])
    print('Atores:', filme['Actors'])
    print('Diretor:', filme['Director'])
    print('Genero:', filme['Genre'])
    print('IMDB:', filme['imdbRating'])
    print('PrÃªmios:', filme['Awards'])
    print('Post:', filme['Poster'])
    print('')

while not sair:
     op = input('Escreva o tÃ­tulo do Filme ou SAIR para fechar: ')
     if op == 'SAIR':
         sair = True
     else:
        filme = requisicao(op)
        if filme['Response'] == 'False':
            print("Filme nÃ£o encontrado :(")
        else:
            printar_detalhes(filme)
```

```python
import requests
import json
import datetime
import time

# https://docs.awesomeapi.com.br/api-de-moedas
requisicao = requests.get('https://economia.awesomeapi.com.br/json/list/USD-BRL/10')
cotacao = json.loads(requisicao.text)

print('#### CotaÃ§Ã£o 7 dias #### ', datetime.datetime.now())

for i in range(len(cotacao)):
    print ('Dolar:', cotacao[i]['bid'])
    print ('Data:', cotacao[i]['create_date'])
    time.sleep(1)
```
